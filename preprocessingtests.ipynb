```python
def rotate_and_save_image(image_path):

    with Image.open(image_path).convert("L") as img:
        base_name = os.path.splitext(os.path.basename(image_path))[0]
        contrasted = PIL.ImageOps.equalize(img)
        denoised = contrasted.filter(PIL.ImageFilter.MedianFilter(size=3))
        denoised.save(image_path)
        mirrored = PIL.ImageOps.mirror(denoised)
        new_image_name = f"{base_name}_mirrored.jpg"
        new_image_path = os.path.join('ece471_data/dataset/screws/train/good/class1', new_image_name)  
        mirrored.save(new_image_path)

        for i in range(1, 4):
            rotated_img = denoised.rotate(90 * i, expand=True)
            new_image_name = f"{base_name}_rotate{i}.jpg"
            new_image_path = os.path.join('ece471_data/dataset/screws/train/good/class1', new_image_name)  
            rotated_img.save(new_image_path)
            print(f"Saved rotated image: {new_image_name}")

        for i in range(1, 4):
            rotated_img = mirrored.rotate(90 * i, expand=True)
            new_image_name = f"{base_name}_mirrored_rotate{i}.jpg"
            new_image_path = os.path.join('ece471_data/dataset/screws/train/good/class1', new_image_name)  
            rotated_img.save(new_image_path)
            print(f"Saved rotated image: {new_image_name}")


cwd = os.getcwd()
cwd = os.path.join(cwd, 'ece471_data/dataset/screws/train/good/class1')

for file_name in os.listdir(cwd):
    if file_name.lower().endswith('.jpg'):
        image_path = os.path.join(cwd, file_name)
        rotate_and_save_image(image_path)
```
